```python
# Web Scraping to get details on car brand sold in each provinces in Canada from Kijiji.Ca

import bs4
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup

my_url = 'https://www.kijiji.ca/b-canada/acura-suv-cars/k0l0?rb=true&dc=true'
uclient = ureq(my_url)
page_html = uclient.read()
uclient.close()

# Html parsing
page_soup = soup(page_html, 'html.parser')

# Grabs each product
Province = page_soup.findAll("a", {"class":"location-selected l1-item"}) # find_all

# To transfer to CSV file
filename = "Acura.csv"
f = open(filename,"w")

headers = "Province,  Quantity\n "
f.write(headers)

for t in Province:
    Province = t.text[20:50]
    Quantity = t.text[78:100]
    

    print("Province:" + Province.replace("   ", ""))
    print("Quantity:"+ Quantity.replace('(','').replace(')','').replace("   ", ""))
    
    # for the csv file
    f.write(Province.replace("   ", "") + "," + Quantity.replace('(','').replace(')','').replace("   ", "") + "\n")
    
f.close()    
```
