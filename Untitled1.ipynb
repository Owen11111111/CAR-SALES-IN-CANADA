```python


# To get the Decrease in Sales YOY from 2019 to 2021

import numpy as np 
# For Visualization
import pandas as pd
import seaborn as sns
# For manipulation
import matplotlib.pyplot as plt
%matplotlib inline 

CanadaCarSales = pd.read_csv('Canadasalesdata.csv')

# Decrease in car sales in Year Over Year from (2019-2021)

# 2019 Total sales
Subset_2019 = CanadaCarSales.loc[CanadaCarSales['Year'].isin([2019])]
Sum2019 = Subset_2019['Sumofsales'].sum()
Sales_2019 = pd.DataFrame({'Year': [2019], 'Sumofsales': [Sum2019]})
Sales_2019

# 2020 Total sales
Subset_2020 = CanadaCarSales.loc[CanadaCarSales['Year'].isin([2020])]
Sum2020 = Subset_2020['Sumofsales'].sum()
Sales_2020 = pd.DataFrame({'Year': [2020], 'Sumofsales': [Sum2020]})
Sales_2020

# 2021 Total sales
Subset_2021 = CanadaCarSales.loc[CanadaCarSales['Year'].isin([2021])]
Sum2021 = Subset_2021['Sumofsales'].sum()
Sales_2021 = pd.DataFrame({'Year': [2021], 'Sumofsales': [Sum2021]})
Sales_2021

# Use these variables to get Total sales for the three years
Sales_2019
Sales_2020
Sales_2021

SOS = pd.concat([Sales_2019,Sales_2020,Sales_2021]).drop_duplicates()

# Add a new column called "Total Sales" with the difference of sales in each row
SOS = SOS.assign(DecreasedSales=SOS['Sumofsales'].diff())

#Add column to have highest sales for YOY
Max_sales = lambda x: SOS['Sumofsales'].max()
SOS['Highest Sales'] =  SOS.apply(Max_sales, axis=1)

# Calculate the percentage of sales for each row
SOS['Sales Percentage'] = (SOS['Sumofsales'] / SOS['Highest Sales']) * 100

# Make the Sales Percentage Column be 2 decimal places
SOS['Sales Percentage'] = SOS['Sales Percentage'].round(2)

# Make the PercentDecreased Column be 2 decimal places
SOS = SOS.assign(PercentDecreased=(SOS['Sales Percentage'].round(2)).diff())

# Add the Percentage symbol Column to the values in the Sales Percentage column
SOS['Sales Percentage'] = SOS['Sales Percentage'].apply(lambda x: str(x) + '%')

# Add the Percentage symbol to the values in the PercentDecreased column
SOS['PercentDecreased'] = SOS['PercentDecreased'].round(2).apply(lambda x: str(x) + '%')

# Print the updated dataframe
SOS

```
